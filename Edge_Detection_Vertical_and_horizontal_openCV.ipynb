```python
import cv2
```

```python
image = cv2.imread("/content/Zebra-in-field-e1653793342190-1024x535.jpg", 0)
```

```python
image.shape  #grayscale image
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image, cmap = 'gray')
```

```python
import numpy as np
kernel =  np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
kernel   #vertical egde detection filter
```

```python
edge_vertical = cv2.filter2D(image, kernel = kernel, ddepth = -1)  #ddepth = -1 is pixel size same as i/p image to o/p image
```

```python
plt.imshow(edge_vertical, cmap = 'gray')
```

 vertical edge are only detected in above 
 now use horizontal

```python
horizontal_filter = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
horizontal_filter
```

```python
edge_horizontal = cv2.filter2D(image, kernel = horizontal_filter , ddepth = -1)
```

```python
plt.imshow(edge_horizontal, cmap = 'gray')
```
