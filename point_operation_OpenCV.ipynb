# Conver to RGB image 

```python
color_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
color_image.shape
```

```python
plt.imshow(color_image)
```

# Point operation

```python
a = 0.5 # a is contrast value
b = 5 # b is brightness value
```

```python
g_x = cv2.convertScaleAbs(image, alpha = a, beta = b)   #point operation gareko
```

```python
plt.imshow(g_x)
```
