```python
import cv2
```

```python
image = cv2.imread('/content/blur-background-lightroom-landscape.jpg', 0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image, cmap = 'gray')
```

```python
import numpy as np
sharp_filter = np.array([[0,-1,0],[-1,6,-1],[0,-1,0]])
```

```python
sharp_filter
```

```python
sharped_image = cv2.filter2D(image, kernel =sharp_filter, ddepth = -1 )
```

```python
plt.imshow(sharped_image, cmap = 'gray')
```

```python
image = cv2.imread('/content/blur-background-lightroom-landscape.jpg', 0)
```
